```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
irish_flower=pd.read_csv('IRIS.csv')
irish_flower.tail()
```

```python
from sklearn import metrics
from sklearn.linear_Lr import LinearRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.Lr_selection import train_test_split
```

```python
irish_flower.describe()
```

```python
irish_flower.isnull().sum()
```

```python
train,test=train_test_split(irish_flower,test_size=0.25)
```

```python
irish_flower["species"].value_counts()
```

```python
sns.FacetGrid(irish_flower,hue="species",height=7).map(plt.scatter,"sepal_length","sepal_width").add_legend()
```

```python
irish_flower.hist()
plt.show()



irish_flower.boxplot()
plt.show()


sns.pairplot(irish_flower, hue='species')
plt.show()
```

```python
test
```

```python
train_X=train[['petal_length','petal_width','sepal_length','sepal_width']]
```

```python
train_y=train[['species']]
```

```python
train_X
```

```python
test_X=test[['petal_length','petal_width','sepal_length','sepal_width']]
test_y=test[['species']]
```

```python
Lr=svm.SVC()
```

```python
Lr.fit(train_X,train_y)
prd=Lr.predict(test_X)
```

```python
val=metrics.accuracy_score(prd,test_y)
val*100
```

```python
Lr=DecisionTreeClassifier()
Lr.fit(train_X,train_y)
prd=Lr.predict(test_X)
```

```python
val=metrics.accuracy_score(prd,test_y)
val*100
```
